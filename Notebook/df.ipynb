```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from collections import defaultdict
from nltk.tokenize import word_tokenize 
from collections import Counter
import re
```

```python
movies = pd.read_csv('Data/cleaned_overview_df.csv')
```

```python
movies_df = movies.drop(['Animation','TV Movie'], axis = 1) 
```

```python
movies_df.iloc[:, 14:]
```

```python
from sklearn.model_selection import train_test_split
X = movies['overview']
y= movies_df['genre']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.3, random_state = 42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df = 10, max_features = 20000, tokenizer = lambda x:x.split(), sublinear_tf = False, ngram_range = (1,4))
X_trained = vectorizer.fit_transform(X_train)
X_tested = vectorizer.transform(X_test)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.dummy import DummyClassifier

dummy_grid = {
    'strategy': ['most_frequent', 'stratified', 'prior', 'uniform', 'constant']
}
dummy = DummyClassifier( random_state = 42)
dummy_gs = GridSearchCV(dummy, param_grid = dummy_grid, cv = 3)
dummy_gs.fit(X_trained, y_train)
print(dummy_gs.best_score_)
```
